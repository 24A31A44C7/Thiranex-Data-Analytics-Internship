```python
# Check missing values in each column
print(df.isnull().sum())
```

```python
# Count duplicate rows
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Check new dataset size
print(df.shape)
```

```python
df.isnull().sum()
```

```python
# Fill missing text values
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    df[col] = df[col].fillna("Unknown")
```

```python
# Fill missing numeric values with mean
numeric_columns = df.select_dtypes(include='number').columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
print(df.isnull().sum())
```

```python
# Remove leading and trailing spaces
for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
# Convert text to proper case
for col in text_columns:
    df[col] = df[col].str.title()
```

```python
df.head()
```

```python
print(df.columns)
```

```python
# Convert Order Date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Month'] = df['Order Date'].dt.month_name()
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
print(df.columns)
```

```python
print("========== SALES REPORT ==========")

print("Total Orders :", len(df))

print("Total Sales :", round(df['Sales'].sum(), 2))

print("Average Sales :", round(df['Sales'].mean(), 2))

print("Maximum Sale :", round(df['Sales'].max(), 2))

print("Minimum Sale :", round(df['Sales'].min(), 2))

print("Total Customers :", df['Customer Name'].nunique())

print("Total Products :", df['Product Name'].nunique())

print("==================================")
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)
```

```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby('Month')['Sales'].sum()

month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

monthly_sales = monthly_sales.reindex(month_order)

plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linewidth=3)

plt.title("Monthly Sales Trend", fontsize=18, weight='bold')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().nlargest(10)

plt.figure(figsize=(12,6))
plt.bar(top_customers.index, top_customers.values)

plt.title("Top 10 Customers", fontsize=16, weight='bold')
plt.xlabel("Customer")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df['Discount'], df['Profit'], alpha=0.5)

plt.title("Discount vs Profit", fontsize=16, weight='bold')
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

```python
ship_mode = df['Ship Mode'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(ship_mode,
        labels=ship_mode.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Shipping Mode Analysis", fontsize=16, weight='bold')

plt.show()
```
